```python
#Finding outliers and the distribution
import seaborn as sns
c = df[['BALANCE','PURCHASES','ONEOFF_PURCHASES','INSTALLMENTS_PURCHASES', 'CASH_ADVANCE', 'CASH_ADVANCE_TRX', 'PURCHASES_TRX', 'CREDIT_LIMIT', 'PAYMENTS','MINIMUM_PAYMENTS', 'PRC_FULL_PAYMENT', 'TENURE']]
p = sns.pairplot(df)
p.fig.suptitle('Colerration')      
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
asg = df[['CASH_ADVANCE', 'BALANCE', 'PURCHASES','ONEOFF_PURCHASES',
          'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE_TRX', 
          'PURCHASES_TRX','CREDIT_LIMIT','PAYMENTS','MINIMUM_PAYMENTS','PRC_FULL_PAYMENT']]
scaler.fit(asg)
dm = scaler.transform(asg)
dm = pd.DataFrame(dm)
dm.columns = ['CASH_ADVANCE', 'BALANCE', 'PURCHASES','ONEOFF_PURCHASES',
          'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE_TRX', 
          'PURCHASES_TRX','CREDIT_LIMIT','PAYMENTS','MINIMUM_PAYMENTS','PRC_FULL_PAYMENT']

df2 = df[['BALANCE_FREQUENCY', 'PURCHASES_FREQUENCY', 'ONEOFF_PURCHASES_FREQUENCY', 
         'PURCHASES_INSTALLMENTS_FREQUENCY', 'CASH_ADVANCE_FREQUENCY','TENURE']]# variables that do not have high variance
df_new = pd.concat([df2, dm], axis=1) #concat two dataframe(dm and df2)
df_new = df_new.set_index(df['CUST_ID'])
df_new.head()
```

```python
#calculating correlation of the variables
df_new.corr(method='pearson')
```

```python
#Using elbow and silhoutte plot to find the best k for clusters
```

```python
from sklearn.cluster import KMeans
Kn = range(1,20)
i = []
for k in Kn:
    model = KMeans(n_clusters=k).fit(df_new)
    i.append([k, model.inertia_])
plt.plot(pd.DataFrame(i)[0], pd.DataFrame(i)[1])
plt.xlabel('k')
plt.ylabel('Sum_of_Squared_Distance')
plt.title('Elbow')
```

```python
from sklearn.metrics import silhouette_score
sse_ = []
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k).fit(df_new)
    sse_.append([k, silhouette_score(df_new, kmeans.labels_)])
plt.plot(pd.DataFrame(sse_)[0], pd.DataFrame(sse_)[1])
plt.xlabel('k')
plt.title('Silhoutte')
plt.show()
```

```python
#Based on elbow and silhoutte, I choose 5 for n_clustering
model = KMeans(n_clusters=5).fit(df_new)
labels = model.predict(df_new)
```

```python
aaa = {}
k = 5
aaa[str(k)] = labels
aaa_best = aaa[str(k)]
counts = np.bincount(aaa_best)
plt.bar(range(k), counts)
```

```python
#Adding Clusters as column to data set
df_new['CLUSTER'] = labels
```

```python
df_new.head()
```

```python
#Finding the centroids of clusters
centers = model.cluster_centers_
centers
```

```python
#putting the centroids for each values
df_cluster = pd.DataFrame(centers)
df_cluster.columns = ['BALANCE_FREQUENCY', 'PURCHASES_FREQUENCY',
       'ONEOFF_PURCHASES_FREQUENCY', 'PURCHASES_INSTALLMENTS_FREQUENCY',
       'CASH_ADVANCE_FREQUENCY', 'TENURE', 'CASH_ADVANCE', 'BALANCE',
       'PURCHASES', 'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES',
       'CASH_ADVANCE_TRX', 'PURCHASES_TRX', 'CREDIT_LIMIT', 'PAYMENTS',
       'MINIMUM_PAYMENTS', 'PRC_FULL_PAYMENT']
df_cluster
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(df_new)
```

```python
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_)
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.xticks(features)
plt.show()
```

```python
#Based on the above plot, chose 2 as n_components
pca = PCA(n_components=2)
w = pca.fit_transform(df_new)
pca1, pca2 = zip(*w)
import matplotlib.colors
cmap = matplotlib.colors.ListedColormap(['red', 'green', 'yellow', 'black', 'orange'])
plt.scatter(pca1, pca2, c=labels.tolist(), cmap=cmap)
```
